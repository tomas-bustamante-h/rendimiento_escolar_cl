```python


# Seleccionar columnas
columnas_seleccionadas = ['Nombre Region',
                          'Clasificación2', 
                          'Matrícula Primer Año',
                            'Valor de arancel', 
                            'Area Conocimiento']

df_filtrado = df[df['Año']==2024][columnas_seleccionadas]


# Mostrar las primeras filas del dataframe filtrado
display(df_filtrado.head())
```

```python
df_filtrado=df_filtrado.dropna()
len(df_filtrado)
```

```python
import pandas as pd
df_final = pd.get_dummies(df_filtrado, 
                          columns=['Nombre Region',
                                   'Clasificación2',
                                   'Area Conocimiento'],
                                   drop_first=False,
                                   dtype=int)

display(df_final.head())
```
